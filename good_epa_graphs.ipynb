```python
from statbotics import Statbotics
from sb_types import TeamYear
from collections import defaultdict
import itertools
from tqdm.auto import tqdm
```

```python
LIMIT = 100
CD_FORMAT = True
```

```python
sb = Statbotics()
```

```python
team_years: dict[str, list[TeamYear]] = defaultdict(list)

for offset in tqdm(itertools.count(0, LIMIT)):
    try:
        team_years_this: list[TeamYear] = sb.get_team_years(limit=LIMIT, offset=offset) # type: ignore
    except UserWarning:
        break
    for ty in team_years_this:
        team_years[f"frc{ty["team"]}"].append(ty)
```

```python
longest_streak_start: dict[str, int] = {}
longest_streak_end: dict[str, int] = {}
longest_streak_length: dict[str, int] = {}
longest_d_streak_start: dict[str, int] = {}
longest_d_streak_end: dict[str, int] = {}
longest_d_streak_length: dict[str, int] = {}
team_name: dict[str, str] = {}

for team, tys in team_years.items():
    tys = sorted(tys, key=lambda ty: ty["year"])
    team_name[team] = tys[-1]["name"]
    start = -2
    end = -2
    longest_start = -2
    longest_end = -2
    d_start = -2
    d_end = -2
    longest_d_start = -2
    longest_d_end = -2
    for i, ty_tuple in enumerate(itertools.pairwise(tys)):
        ty1, ty2 = ty_tuple
        if ty1["epa"]["norm"] <= ty2["epa"]["norm"]:
            if end == i:
                end = i + 1
            else:
                start = i
                end = i + 1
        if end - start >= longest_end - longest_start:
            longest_start = start
            longest_end = end
        if ty1["epa"]["norm"] >= ty2["epa"]["norm"]:
            if d_end == i:
                d_end = i + 1
            else:
                d_start = i
                d_end = i + 1
        if d_end - d_start >= longest_d_end - longest_d_start:
            longest_d_start = d_start
            longest_d_end = d_end
    if longest_start == -2:
        longest_streak_start[team] = 0
        longest_streak_end[team] = 0
        longest_streak_length[team] = 0
    else:
        longest_streak_start[team] = tys[longest_start]["year"]
        longest_streak_end[team] = tys[longest_end]["year"]
        longest_streak_length[team] = longest_end - longest_start + 1
    if longest_d_start == -2:
        longest_d_streak_start[team] = 0
        longest_d_streak_end[team] = 0
        longest_d_streak_length[team] = 0
    else:
        longest_d_streak_start[team] = tys[longest_d_start]["year"]
        longest_d_streak_end[team] = tys[longest_d_end]["year"]
        longest_d_streak_length[team] = longest_d_end - longest_d_start + 1
```

```python
if CD_FORMAT:
    print("| Team | Improvement Streak Length | Start Year | End Year |")
    print("| --- | --- | --- | --- |")

for team in sorted(longest_streak_length, key=lambda t: longest_streak_length[t], reverse=True)[:20]:
    num = team[3:]
    start = longest_streak_start[team]
    end = longest_streak_end[team]
    length = longest_streak_length[team]
    if CD_FORMAT:
        print(f"| [{num}](https://thebluealliance.com/team/{num}) | {length} | [{start}](https://thebluealliance.com/team/{num}/{start}) | [{end}](https://thebluealliance.com/team/{num}/{end}) |")
    else:
        print(f"{team}: {length} ({start}-{end})")
```

```python
if CD_FORMAT:
    print("| Team | Decline Streak Length | Start Year | End Year |")
    print("| --- | --- | --- | --- |")

for team in sorted(longest_d_streak_length, key=lambda t: longest_d_streak_length[t], reverse=True)[:20]:
    num = team[3:]
    start = longest_d_streak_start[team]
    end = longest_d_streak_end[team]
    length = longest_d_streak_length[team]
    if CD_FORMAT:
        print(f"| [{num}](https://thebluealliance.com/team/{num}) | {length} | [{start}](https://thebluealliance.com/team/{num}/{start}) | [{end}](https://thebluealliance.com/team/{num}/{end}) |")
    else:
        print(f"{team}: {length} ({start}-{end})")
```

```python
if "get_ipython" not in globals():
    import json
    import pathlib

    output_folder = pathlib.Path(__file__).parent / "build"

    config = {
        "columns": [
            {
                "name": "team",
                "type": "team",
                "displayName": "Team",
            },
            {
                "name": "length",
                "type": "integer",
                "displayName": "Streak Length",
            },
            {
                "name": "start",
                "type": "integer",
                "displayName": "Start Year",
            },
            {
                "name": "end",
                "type": "event",
                "displayName": "End Year",
            },
            {
                "name": "active",
                "type": "boolean",
                "displayName": "Active",
            },
        ],
        "defaultSort": [{"column": "length", "direction": -1}],
        "primaryKey": "team",
    }

    with (output_folder / "increasing-epa-streaks.config.json").open("w") as f:
        json.dump({**config, "title": "Increasing EPA Streaks"}, f, indent=None)
    with (output_folder / "increasing-epa-streaks.json").open("w") as f:
        json.dump(
            [
                {
                    "team": team_key,
                    "length": longest_streak_length[team_key],
                    "start": longest_streak_start[team_key],
                    "end": longest_streak_end[team_key],
                    "active": longest_streak_end[team_key] == 2025,
                }
                for team_key in longest_streak_length.keys()
            ],
            f,
            indent=None,
        )
    with (output_folder / "decreasing-epa-streaks.config.json").open("w") as f:
        json.dump(
            {
                **config,
                "title": "Decreasing EPA Streaks",
            },
            f,
            indent=None,
        )
    with (output_folder / "decreasing-epa-streaks.json").open("w") as f:
        json.dump(
            [
                {
                    "team": team_key,
                    "length": longest_d_streak_length[team_key],
                    "start": longest_d_streak_start[team_key],
                    "end": longest_d_streak_end[team_key],
                    "active": longest_d_streak_end[team_key] == 2025,
                }
                for team_key in longest_d_streak_length.keys()
            ],
            f,
            indent=None,
        )
```
